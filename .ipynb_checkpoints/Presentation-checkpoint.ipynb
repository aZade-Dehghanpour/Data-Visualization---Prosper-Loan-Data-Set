## Dataset
Dataset consist of **113,937** entries on loans borrowed through 'Prosper' peer-to-peer lending marketplace. Each entry (row) has **81** features (fields/columns) that describe certain attributes of the Borrower (Member),  borrower's loan request(Listing), Lender(Investor), and the Loan itself including loan amount, borrower rate (or interest rate), current loan status, borrower income, and many others.
Further elaboration on variables can be found at [Prosper API Definition Page](https://www.prosper.com/Downloads/Services/Documentation/ProsperAPI_Objects_Details.html).


## Main Areas of Interest in Dataset
 - Makeup of Prosper customers (Borrowers)in terms of the level of their income (income range and monthly income), available credit and credit grade.
 - Categories in which the members apply for loans (business, home improvement,...), size of the loan they usually borrow for the stated purpose and what repayment schedule they desire most ( 12,36 or 60 months) 
 - Relationship between certain features of members (monthly income, available credit, debt to income ratio) and Annual Percentage Rate(APR) they recieve.


## Dataset Features of Interest
 - Income Range
 - Stated Monthly Income
 - Debt to Income Ratio
 - Available Bankcard Credit
 - Credit Grade
 - Loan Original Amount
 - Term
 - Borrower APR




## Limitation
- **Listing Category**
A significant portion of loans are listed under other or Not available. Therefore the analysis falls short of accurately make a statement on what the loans are borrowed for.

- **Credit Grade**
For 85% of the entries this valuable indicator is not available. Though quite an important and interesting feature, lack of enough entries caused a degree of descrepency.

## Listing Category
**51%** of loans are taken for **Debt Consolidation**. Unfortunately 25% of entries offer un-informative data stating 'Not Available' and 'Other' for Listing Category. It should be decided later on how to handle this issue.

**The category of the listing that the borrower selected when posting their listing**<br>
0 - Not Available, 1 - Debt Consolidation, 2 - Home Improvement, 3 - Business, 4 - Personal Loan, 5 - Student Use, 6 - Auto, 7- Other, 8 - Baby&Adoption, 9 - Boat, 10 - Cosmetic Procedure, 11 - Engagement Ring, 12 - Green Loans, 13 - Household Expenses, 14 - Large Purchases, 15 - Medical/Dental, 16 - Motorcycle, 17 - RV, 18 - Taxes, 19 - Vacation, 20 - Wedding Loans


````python
clean_data.rename(columns = {'ListingCategory (numeric)': 'ListingCategory'}, inplace = True)
````

````python
clean_data.ListingCategory.isna().sum()
````

````python
listing_category_dict = {}
listing_category_list = ['Not Available', 'Debt Consolidation', 'Home Improvement', 'Business','Personal Loan', 'Student Use', 'Auto', 'Other', 'Baby&Adoption', 'Boat', 'Cosmetic Procedure', 'Engagement Ring', 'Green Loans', 'Household Expenses', 'Large Purchases', 'Medical/Dental', 'Motorcycle', 'RV', 'Taxes', 'Vacation', 'Wedding Loans']
for n,lc in enumerate(listing_category_list):
    listing_category_dict[n] = lc
listing_category_dict
````

````python
#replace numeric value with name of the category
for i in range(len(listing_category_dict)):
    rows_to_change = clean_data[clean_data.ListingCategory == i].index 
    for j in rows_to_change:
        clean_data.loc[j,'ListingCategory'] = listing_category_dict[i]
````

````python
clean_data.ListingCategory.unique()
````

# Bivariant Exploration

In previous steps, **Income Range** appeared to be viable feature based on which the data can be categorized and further explored.  There are **no NaN value** and the categories are informative enough. However, categories such as'Not Displayed' does not offer much explanation and the categories \\$0' and \\$1-24,999. <br>
**85%** of members fall into **4 categories** ( \\$25,000-49,999,\\$50,000-74,999, \\$75,000-99,999, \\$100,000+). To ensure these for categories have enough share in the loans borrowed, in the next step, the distribution of loans(percentage of total amount) across categories of ```IncomeRange``` is explored. If the criteria of being a majory group is met, focus data will be data related to these four categories only. 

## Majority Groups in Dataset (Focus Data)
**Four income range categories** (\\$25,000-49,999,\\$50,000-74,999,\\$75,000-99,999,\\$100,000+) are the majority groups by number of loans **85%** of all the loans, and volum **~92%** of the total volume borrowed.<br> 
***Members belonging to these four categories were the focus of this analysis***

````python
income_categories = ['$0', '$1-24,999','$25,000-49,999','$50,000-74,999','$75,000-99,999', '$100,000+','Not employed', 'Not displayed']
income_type = pd.api.types.CategoricalDtype(ordered = True, categories = income_categories)
clean_data.IncomeRange = clean_data.IncomeRange.astype(income_type)
````

````python
total_loan_per_income_category = clean_data.groupby('IncomeRange').LoanOriginalAmount.sum().reset_index(name = 'TotalLoan')
total_loan_per_income_category
````

````python
#plt.style.context('fivethirtyeight')
plt.figure(figsize = (18,10))
plt.style.context('fivethirtyeight')

plt.subplot(121)
sb.countplot(data = clean_data, x = 'IncomeRange', color = base_color)
plt.xticks(size = 16,rotation = '90')
plt.yticks(size = 16)
plt.xlabel('Income Range' , size = 18)
plt.ylabel('Number of Loans (%of Total)' , size = 20)
plt.title('Distribution of Loans across Income Range Categories',size = 20)

IncomeRange_count = clean_data.IncomeRange.value_counts()
xtick_loc, xtick_label = plt.xticks()
all_IncomeRange_count = clean_data.IncomeRange.shape[0] - clean_data.IncomeRange.isna().sum()
for xt_loc, xt_lab in zip(xtick_loc, xtick_label):
    count_ony = IncomeRange_count[xt_lab.get_text()]
    plt.text(xt_loc,count_ony, '{:.2f}%'.format(100*count_ony/all_IncomeRange_count), ha = 'center',size = 16);

plt.subplot(122)
patches, texts, autotexts =plt.pie(total_loan_per_income_category.TotalLoan,startangle = 90, counterclock = False, autopct='%1.1f%%',pctdistance=1.1, labeldistance=2,wedgeprops = {'width' : 0.4}, textprops={'fontsize': 16})
plt.legend(labels = total_loan_per_income_category.IncomeRange, loc='center left', bbox_to_anchor=(1.2, 0.6), fontsize = 16);
plt.title('Distribution of Total Loan Volume across Income Range Categories ', size = 20, loc='center')
plt.setp(autotexts[0], size=16, rotation = '90')
plt.setp(autotexts[1], size=16, rotation = '90')
plt.setp(autotexts[6], size=16, rotation = '90')
plt.setp(autotexts[7], size=16, rotation = '90');
````

>**Four income range categories** (\\$25,000-49,999,\\$50,000-74,999,\\$75,000-99,999,\\$100,000+) account for **~92%** of the amount of money borrowed. Therefore,in the steps to come the focus of this analysis will be on these categories only.
